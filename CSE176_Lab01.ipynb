```python
from transformers import pipeline
clf = pipeline("image-classification")
clf("https://ucmercedcse176.github.io/files/resource/cat.jpeg")
```

```python
from PIL import Image
from transformers import pipeline

vqa_pipeline = pipeline("visual-question-answering")

image =  Image.open("/content/cat.jpeg")
question = "Is there a cat?"

vqa_pipeline(image, question, top_k=1)
#[{'score': 0.9998154044151306, 'answer': 'yes'}]
```

```python
from transformers import pipeline

captioner = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
captioner("https://ucmercedcse176.github.io/files/resource/cat.jpeg")
## [{'generated_text': 'two birds are standing next to each other '}]
```

```python
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler
import torch

model_id = "stabilityai/stable-diffusion-2"
scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")
pipe = StableDiffusionPipeline.from_pretrained(model_id, scheduler=scheduler, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "a cat looking up at the sky"
image = pipe(prompt).images[0]

image.save("Generated_Image.png")
```
